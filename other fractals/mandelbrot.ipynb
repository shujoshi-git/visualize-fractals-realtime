```python
import numpy as np
import matplotlib.pyplot as plt

# Function to check if a point is in the Mandelbrot set
def mandelbrot(c, max_iter=256):
    z = 0
    for i in range(max_iter):
        z = z**2 + c
        if abs(z) > 2:  # If the point escapes the set
            return i
    return max_iter  # If the point does not escape

# Function to generate and plot the Mandelbrot set
def plot_mandelbrot(x_min, x_max, y_min, y_max, width, height, max_iter=256):
    # Generate a grid of complex numbers in the specified range
    x, y = np.linspace(x_min, x_max, width), np.linspace(y_min, y_max, height)
    X, Y = np.meshgrid(x, y)
    C = X + 1j * Y  # Create the complex grid of numbers
    
    # Initialize an array to store the iteration counts for each point
    Z = np.zeros(C.shape, dtype=int)
    
    # Iterate through each point in the grid
    for i in range(height):
        for j in range(width):
            Z[i, j] = mandelbrot(C[i, j], max_iter)  # Store iteration count

    # Plot the Mandelbrot set
    plt.figure(figsize=(10, 10))
    plt.imshow(Z, cmap='Blues', extent=(x_min, x_max, y_min, y_max))  # Use 'hot' colormap for better visualization
    plt.colorbar()
    plt.title(f"Mandelbrot Set (Max Iterations = {max_iter})")
    plt.xlabel('Re')
    plt.ylabel('Im')
    plt.show()

# Example usage: Plot the Mandelbrot set with specified boundaries and resolution
x_min, x_max = -2.0, 1.0  # Real axis range
y_min, y_max = -1.5, 1.5  # Imaginary axis range
width, height = 800, 800  # Resolution of the plot
max_iter = 256  # Maximum number of iterations

plot_mandelbrot(x_min, x_max, y_min, y_max, width, height, max_iter)
```
