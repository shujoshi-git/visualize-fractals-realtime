```python
import matplotlib.pyplot as plt
import numpy as np

def custom_koch_curve(points, iterations, n):
    if iterations == 0:
        return points
    
    new_points = []
    for i in range(len(points) - 1):
        p1, p2 = points[i], points[i + 1]
        
        # Compute segment length
        length = np.linalg.norm(p2 - p1)
        segment_fraction = 1 / (4 ** n)
        triangle_side = segment_fraction
        
        # Compute the new points
        a = p1
        b = p1 + (p2 - p1) * (0.5 - segment_fraction / 2)
        d = p1 + (p2 - p1) * (0.5 + segment_fraction / 2)
        
        # Compute the peak of the equilateral triangle
        theta = np.pi / 3  # 60 degrees
        direction = (d - b) / np.linalg.norm(d - b)
        rot_matrix = np.array([[np.cos(theta), -np.sin(theta)], [np.sin(theta), np.cos(theta)]] if np.cross(direction, [0, 1]) >= 0 else [[np.cos(-theta), -np.sin(-theta)], [np.sin(-theta), np.cos(-theta)]])
        peak = np.dot(rot_matrix, (d - b)) + b
        
        # Append new points
        new_points.extend([a, b, peak, d])
    
    new_points.append(points[-1])  # Add the last point
    return custom_koch_curve(new_points, iterations + 1, n + 1)

def plot_custom_koch_curve(iterations):
    points = np.array([[0, 0], [1, 0]])  # Start with a line segment [0,1]
    curve = custom_koch_curve(points, iterations, 1)
    
    curve = np.array(curve)
    plt.figure(figsize=(8, 2))
    plt.plot(curve[:, 0], curve[:, 1], 'b-')
    plt.axis('equal')
    plt.title(f'Custom Koch-Like Curve (Iteration {iterations})')
    plt.show()

# Example usage
plot_custom_koch_curve(4)
```
